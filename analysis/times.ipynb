```python
import numpy as np 
import pandas as pd 
import pylab as plt
```

```python
data = pd.read_csv("times.csv", index_col=0)
data.head()
```

```python
container = dict()
c = 0
for col in data.columns:
    for cell in data[col]:
        container[c] = {
            "value": cell,
            "group": "E" if "E" in col else "H" if "H" in col else "S" 
        }
        c += 1
df = pd.DataFrame(container).transpose()
df.head()
```

```python
df.to_csv("times_update.csv")
```
