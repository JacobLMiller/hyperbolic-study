```python
import json 
import graph_tool.all as gt
from find_answers import *
```

```python
with open("../src/application/static/data/test-questions.json", "r") as fdata:
    data = json.load(fdata)
questions = data["questions"]
```

```python
new_questions = [[q, None, None] for q in questions]
new_questions[0]
```

```python
task_dict = {
    "T1": lambda G, name : append_question_node(count_degree, G, random.randint(2,8), name, "T1"), 
    "T2": lambda G,name : append_question_node(common_neighbors, G, random.randint(1,8), name, "T2"),
    "T3": lambda G,name : append_question_node(shortest_path, G, random.randint(3,7), name, "T3"), 
    "T4": lambda G,name : estimate_size(G,name),
    "T5": lambda G,name : find_mode(G,name), 
    "T6": lambda G,name : find_bfs(G,name)
}
```

```python
for q_arr in new_questions: 
    old_q = q_arr[0]
    q_func = task_dict[ old_q["q_type"] ]

    with open(f"../src/application/static/data/{old_q['graph']['graph_id']}.json", "r") as fdata:
        gdata = json.load(fdata)
    G = gt.Graph(directed=False)
    G.add_vertex(len(gdata['nodes']))
    G.add_edge_list([(e["source"], e["target"]) for e in gdata["links"]])
    name = old_q['graph']['graph_id']

    for i in range(1,3):
        q_arr[i] = q_func(G,name)
        q_arr[i]["question_id"] = f"{old_q['question_id']}_{i}"
    old_q['question_id'] = f"{old_q['question_id']}_0"
```
